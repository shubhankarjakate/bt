```python
import hashlib


# Function to generate SHA-256 hash
def calculate_hash(data):

    return hashlib.sha256(data.encode()).hexdigest()


# Function to compute Merkle Root
def compute_merkle_root(transactions):

    # Step 1: Hash all transactions
    transaction_hashes = [
        calculate_hash(tx)
        for tx in transactions
    ]

    print("\nTransaction Hashes:")
    for h in transaction_hashes:
        print(h)

    # Step 2: Build Merkle Tree
    while len(transaction_hashes) > 1:

        temp_list = []

        # Process pairs
        for i in range(0, len(transaction_hashes), 2):

            left = transaction_hashes[i]

            # If odd number of hashes
            if i + 1 < len(transaction_hashes):
                right = transaction_hashes[i + 1]
            else:
                right = left

            # Combine and hash again
            combined_hash = calculate_hash(left + right)

            temp_list.append(combined_hash)

        transaction_hashes = temp_list

    return transaction_hashes[0]


# Sample Transactions
transactions = [
    "Alice pays Bob 5 BTC",
    "Bob pays Charlie 2 BTC",
    "Charlie pays David 1 BTC",
    "David pays Emma 0.5 BTC"
]


# Compute Merkle Root
merkle_root = compute_merkle_root(transactions)

print("\nMerkle Root:")
print(merkle_root)
```
