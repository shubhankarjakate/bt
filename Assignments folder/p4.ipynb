```python
# Wallet Class
class Wallet:

    def __init__(self, owner, balance):

        self.owner = owner
        self.balance = balance

    # Send money function
    def send_money(self, receiver, amount):

        print(f"\n{self.owner} tries to send {amount} coins to {receiver.owner}")

        # Check balance
        if self.balance >= amount:

            self.balance -= amount
            receiver.balance += amount

            print("Transaction Successful")

        else:

            print("Transaction Failed: Insufficient Balance")


# Create wallets
wallet_A = Wallet("Alice", 100)
wallet_B = Wallet("Bob", 50)

# Initial Balances
print("Initial Balances")
print("Alice:", wallet_A.balance)
print("Bob:", wallet_B.balance)

# Valid Transaction
wallet_A.send_money(wallet_B, 40)

# Balances After Transaction
print("\nBalances After Transaction")
print("Alice:", wallet_A.balance)
print("Bob:", wallet_B.balance)

# Double Spending Attempt
wallet_A.send_money(wallet_B, 80)

# Final Balances
print("\nFinal Balances")
print("Alice:", wallet_A.balance)
print("Bob:", wallet_B.balance)
```
